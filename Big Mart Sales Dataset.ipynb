# Data Cleaning

Frome the above basic analysis we get know that there are some null values present in two columns such as 'Item_Weight' and 'Outlet_Size'.

<h4>3.1 Filling null values in the 'Item_Weight' column</h4>

```python
#Checking based on 'Item_Identifier' can we fill the null values or not
df[df['Item_Identifier']=='FDA15'] # checking with first row 'Item_Identifier' value 
```

From the above output we can clearly see that based on the 'Item_Identifier' we are having a similar 'Item_Weight'.
So, by taking the average of the 'Item_Weight' of each 'Item_Identifier' we can fill the null values present with in the 'Item_Weight' column.

```python
#Calculating mean of Item_Weight for each 'Item_Identifier' and filling NaN with the mean value
df['Item_Weight']=df.groupby('Item_Identifier')['Item_Weight'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['Item_Weight'].isnull().sum()
```

After filling with the above condition we still have 4 null values in the 'Item_Weight' column.

```python
df[df['Item_Weight'].isnull()]
```

Lets fill above null values based on the each 'Item_Type' average weight.

```python
#Calculating mean of Item_Weight for each 'Item_Type' and filling NaN with the mean value
df['Item_Weight']=df.groupby('Item_Type')['Item_Weight'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['Item_Weight'].isnull().sum()
```

No null values present in the 'Item_Weight' column.

<h4>3.2 Filling null values in the 'Outlet_Size' column</h4>

```python
df['Outlet_Size'].value_counts(dropna=False)
```

From the above output we can clearly see that ther are 2410 null values are present in the 'Outlet_Size' column

```python
pd.crosstab(index=df['Outlet_Size'], columns=df['Outlet_Type'])
```

From the above output we can clearly see that 'Grocery Store' Outlet_Type contains only 'Small' Outlet_Size.
So, let's fill the Outlet_Size column with 'Small' as a value where its 'Outlet_Type' is 'Grocery Store'.

```python
#Checking for null values present in the 'Outlet_Size' column for 'Grocery Store' Outlet Type
df[df['Outlet_Type']=='Grocery Store'].isnull().sum() 
```

```python
#Filling null values with 'Small' Outlet Size For 'Grocery Store' outlet type
df['Outlet_Size']=df['Outlet_Size'].fillna(df['Outlet_Type'].map({'Grocery Store':'Small'}))
```

In the above output Supermarket Type2 and Supermarket Type3 are also having only midium 'Outlet_Size'. Let's check can we fill or not.

```python
#Checking for null values present in the 'Outlet_Size' column for 'Supermarket Type2' Outlet Type
df[df['Outlet_Type']=='Supermarket Type2'].isnull().sum()
```

```python
#Checking for null values present in the 'Outlet_Size' column for 'Supermarket Type3' Outlet Type
df[df['Outlet_Type']=='Supermarket Type3'].isnull().sum()
```

From the above two analysis we can conclude that we can't fill 'Outlet_Size' column null values with the correspondig 'Outlet_Type'.

```python
df['Outlet_Size'].value_counts(dropna=False)
```

Still we are having null values in the 'Outlet_Size'column.Let's compare 'Outlet_Size' and 'Outlet_Location_Type'.

```python
pd.crosstab(index=df['Outlet_Size'], columns=df['Outlet_Location_Type'])
```

We can see that Tier2 'Outlet_Location_Type' having only 'Small' Outlet_Size. So, let's fill the null values in the 'Outlet_Size' column with 'Small' value where its 'Outlet_Location_Type' is 'Tier2'.

```python
df['Outlet_Size']=df['Outlet_Size'].fillna(df['Outlet_Location_Type'].map({'Tier 2':'Small'}))
```

```python
df['Outlet_Size'].value_counts(dropna=False)
```

```python
df.isnull().sum()
```

No null values are present in the dataframe.

<h5>Insights:</h5>
From the above graph we can't predict any analysis. So let's add one column to our dataframe with a price range of Item_MRP.

```python
#Defining a function for Price_Range column
def categorize(Price_Range):
    if Price_Range <= 30:
        return "0-30" 
    elif Price_Range <=70 :
        return "31-70"
    elif Price_Range <=110 :
        return "71-110" 
    elif Price_Range <=150 :
        return "111-150"
    elif Price_Range <=190 :
        return "151-190"
    else :
        return "190+"
 
df["Price_Range"] = df["Item_MRP"].apply(categorize) #Adding new column 'Price_Range' to the dataframe
```

```python
price_range_df=pd.DataFrame(df['Price_Range'].value_counts().reset_index(name='Number_of_products'))
price_range_df
```

```python
custom_colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#c2c2f0']
plt.pie(price_range_df['Number_of_products'],labels=price_range_df['Price_Range'],autopct='%1.2f%%',colors=custom_colors,shadow=True)
plt.show()
```

<h5>Insights:</h5>
From the above analysis, Item_MRP with in the price range '151-190'and '190+' are the most 'Item_Type' products are present in our dataset.
